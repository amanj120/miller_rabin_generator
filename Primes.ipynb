```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
primes = [2, 3, 5, 7, 11]
for i in range(13, 1<<16, 2):
    found = True
    for p in primes:
        if i%p == 0:
            found = False
            break
    if found == True:
        primes.append(i)
primes
```

```python
prod = 1;
products = [];
for p in primes:
    products.append(prod * p)
    prod *= p
# products
```

```python
mods = 100
fib1 = 0;
fib2 = 1;
fibsmod = [0, 1]
for i in range(1<<20):
    tmp = fib1 + fib2;
    fib1 = fib2 % mods;
    fib2 = tmp % mods;
    fibsmod.append(tmp % mods)
```

```python
counts = np.zeros(mods, dtype=np.int32)
for i in fibsmod:
    counts[i] += 1
```

```python
plt.figure(figsize=(20,10))
plt.bar(list(range(mods)), counts)
```
